```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Creating a Custom DataFrame
data = pd.DataFrame({
    'Income': [25000, 50000, 75000, 100000, 125000, 150000, 175000, 200000, 225000, 250000],
    'Age': [22, 25, 47, 52, 23, 57, 48, 33, 40, 29],
    'SpendingScore': [39, 81, 6, 77, 40, 88, 55, 50, 49, 45]
})

# Step 2: Standardizing the Data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Step 3: Applying PCA (Reducing to 2 Principal Components)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

# Step 4: Creating a DataFrame for PCA Results
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
print("\nPCA Transformed Data:\n", pca_df)

# Step 5: Explained Variance Ratio
explained_variance = pca.explained_variance_ratio_
print("\nExplained Variance Ratio:", explained_variance)

# Step 6: Visualizing PCA Results
plt.figure(figsize=(8, 5))
plt.scatter(pca_df['PC1'], pca_df['PC2'], c='blue', edgecolors='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.grid()
plt.show()
```

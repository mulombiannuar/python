```python
# load libraries : Requires scikit-learn, numpy, matplotlib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
# define dataset for the model
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 10, 20])
```

```python
# create pipeline: polynomial features (degree=3) then linear regression
model = make_pipeline(
    PolynomialFeatures(degree=3, include_bias=False), LinearRegression()
)
```

```python
# fit model
model.fit(x, y)
```

```python
# coefficients: retrieve from the last step
lin = model.named_steps["linearregression"]
coefs = lin.coef_  
intercept = lin.intercept_
```

```python
print("Intercept:", intercept)
print("Coefficients (for x, x^2, x^3):", coefs)
```

```python
# predictions for plotting
x_plot = np.linspace(1, 5, 200).reshape(-1, 1)
y_plot = model.predict(x_plot)
```

```python
# Plot data and fitted curve
plt.figure(figsize=(6,4))
plt.scatter(x, y, label="Data points")
plt.plot(x_plot, y_plot, label="Cubic polynomial fit", linewidth=1.5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Cubic Polynomial Regression (scikit-learn)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
